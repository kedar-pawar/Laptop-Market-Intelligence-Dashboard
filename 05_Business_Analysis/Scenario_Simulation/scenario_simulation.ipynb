## SCENARIO 2 — RAM Upgrade Strategy

```python
df_s2 = df_scaled.copy()
```

#### - All products get +4GB RAM.

```python
df_s2['ram_gb_upgrade'] = df_s2['ram_gb'] + 4
```

#### - Rescaling featues

```python
from sklearn.preprocessing import MinMaxScaler

features_upgrade = ['cpu_tier', 'clock_speed_ghz', 'ram_gb_upgrade', 'ssd_gb']

scaler = MinMaxScaler()

df_scaled_features = df_s2.copy()
df_scaled_features[features_upgrade] = scaler.fit_transform(
    df_s2[features_upgrade]
)
```

#### - Recalculate Performance Score

```python
df_scaled_features['performance_score_upgrade'] = (
    df_scaled_features['cpu_tier'] * 0.4 +
    df_scaled_features['clock_speed_ghz'] * 0.3 +
    df_scaled_features['ram_gb_upgrade'] * 0.2 +
    df_scaled_features['ssd_gb'] * 0.1
)
```

#### Recalculate Value Index

```python
df_scaled_features['value_index_upgrade'] = (
    df_scaled_features['performance_score_upgrade'] /
    df_scaled_features['Price']
)
```

#### Measure Improvement %

```python
df_scaled_features['value_index_gain_%'] = (
    (df_scaled_features['value_index_upgrade'] - df_scaled_features['value_index'])
    / df_scaled_features['value_index']
) * 100
```

#### Brand-Level Gain

```python
ram_upgrade_impact = (
    df_scaled_features.groupby('Company')['value_index_gain_%']
    .mean()
    .sort_values(ascending=False)
)

ram_upgrade_impact
```
