```python
import pandas as pd
```

```python
df_scaled = pd.read_csv(
    "../Value_Index_Design/value_index_model_ready.csv"
)
```

## Simulating Volume

## Unit sold

```python
df_scaled['estimated_units_sold'] = (
    df_scaled['value_index'] * 100000
).round()
```

## Revenue Estimation

```python
df_scaled['estimated_revenue'] = (
    df_scaled['estimated_units_sold'] * df_scaled['Price']
)
```

## Brand Revenue Contribution

```python
unit_sold = (
    df_scaled.groupby('Company')['estimated_units_sold']
    .sum()
)
unit_sold
```

```python
total_units = df_scaled['estimated_units_sold'].sum()

brand_units = (
    df_scaled.groupby('Company')['estimated_units_sold']
    .sum()
)

market_share = (brand_units / total_units) * 100
market_share
```

```python
brand_revenue = (
    df_scaled.groupby('Company')['estimated_revenue']
    .sum()
    .sort_values(ascending=False)
)
for i in brand_revenue.values:
    print(i)
```

```python
f_mix = pd.merge(
    brand_revenue,market_share,
    left_index=True,
    right_index=True
)
mix = pd.merge(
    unit_sold,f_mix,
    left_index=True,
    right_index=True
)
for i in mix['estimated_units_sold_y']:
    print(i)
mix
```

```python
mix.to_excel(
    "brand_revenue_simulation.xlsx"
)
```
