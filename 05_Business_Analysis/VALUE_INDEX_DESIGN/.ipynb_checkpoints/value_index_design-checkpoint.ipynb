```python
import pandas as pd
```

```python
df=pd.read_csv("../../02_Data_Management/04_Processed_Data/analytics_dataset.csv")
```

```python
df.columns
```

```python
df_scaled = df.copy()
df_scaled['value_index'] = df_scaled['Performance_Score'] / df_scaled['Price']
df_scaled['value_index']
```

## 2 Create Business Buckets

```python
df_scaled['value_segment'] = pd.qcut(
    df_scaled['value_index'],
    q=3,
    labels=['Low Value', 'Balanced', 'High Value']
)
```

```python
df_scaled.to_excel("Value_Index_Output.xlsx", index=False)
```

```python
df_scaled.to_csv(
    "value_index_model_ready.csv",
    index=False
) # in order to use it in different folder and file
```
