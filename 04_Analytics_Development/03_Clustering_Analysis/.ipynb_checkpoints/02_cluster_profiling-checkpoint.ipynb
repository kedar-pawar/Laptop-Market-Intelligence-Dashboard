```python
import pandas as pd
```

## 02_A_Hardware_Based_Cluster_Profiling

#### 1.  Load dataset

```python
df_hardware=pd.read_csv ('hardware_clustered_laptops.csv')
df_hardware.columns
```

#### 2. Compute Cluster Means

```python
hardwar_cluster_profile = df_hardware.groupby("cluster").agg({
    "Price": "mean",
    "ram_gb": "mean",
    "total_storage_gb": "mean",
    "cpu_tier_numeric": "mean",
    "ssd_indicator": "mean"
}).round(2)

hardwar_cluster_profile
```

#### 3. Add Distribution Insights

##### CPU Tier Distribution

```python
pd.crosstab(df_hardware["cluster"], df_hardware["cpu_tier_numeric"], normalize="index")
```

##### Brand Dominance

```python
pd.crosstab(df_hardware["cluster"], df_hardware["Company"], normalize="index")
```

##### Touchscreen %

```python
df_hardware.groupby("cluster")["is_touchscreen"].mean()
```

#### 4 Build Profile Table

```python
hardware_profile=(
        df_hardware.groupby("cluster")
        .agg({
            "Price": "mean",
            "ram_gb": "mean",
            "total_storage_gb": "mean",
            "cpu_tier_numeric": "mean",
            "is_touchscreen": "mean"
        })
        .round(2)
        .reset_index()
)
hardware_profile.columns=[
        "Clusters",
        "Avg Price",
        "Avg Ram(GB)",
        "Avg Storage (GB)",
        "Avg CPU_tier_Numeric",
        "Avg TouchScreen"
    ]
hardware_profile
```

```python
hardware_profile["Segment Name"] = [
    "Premium Touch Ultrabooks",
    "Storage-Focused Productivity",
    "Entry-Level Budget"
]
hardware_profile = hardware_profile.sort_values("Avg Price")
hardware_profile
```

#### conclusion:

- Although Cluster 1 has slightly stronger raw specifications (RAM and CPU tier), Cluster 0 commands higher pricing due to premium features such as touchscreen and likely lightweight design.

- This indicates that in this dataset, user experience and design features contribute significantly to price positioning, sometimes more than raw hardware capacity.

##### Cluster 0 — Premium Touch Ultrabooks



- Highest average price (~₹76K)

- 100% touchscreen penetration

- Strong CPU tier

- SSD-dominant configuration

- This segment represents premium ultrabooks emphasizing portability, touchscreen experience, and balanced high-end performance.

##### Cluster 1 — Storage-Focused Productivity



- Mid-range price (~₹67K)

- Highest average storage capacity

- Strong CPU tier

- Minimal touchscreen presence

- This cluster appears optimized for productivity users requiring large storage capacity rather than premium design features.

##### Cluster 2 — Entry-Level Budget

- Lowest average price (~₹49K)

- Lowest RAM and CPU tier

- No touchscreen presence

- Moderate SSD penetration

- This segment represents cost-sensitive consumers prioritizing affordability over performance and premium features.

## 02_B_Market_Based_Cluster_Profiling

#### 1. Load Dataset

```python
df_market=pd.read_csv ('Market_clustered_laptops.csv')
df_market.columns
```

#### 2. Compute Cluster Means

```python
market_cluster_profile = (
    df_market.groupby("cluster")
    .agg({
        "Price": "mean",
        "ram_gb": "mean",
        "total_storage_gb": "mean",
        "cpu_tier_numeric": "mean",
        "ssd_indicator": "mean",
        "is_touchscreen": "mean"
    })
    .round(2)
    .reset_index()
)

market_cluster_profile
```

#### 3 Add Distribution Insights

##### CPU Tier Distribution

```python
pd.crosstab(
    df_market["cluster"],
    df_market["cpu_tier_numeric"],
    normalize="index"
)
```

##### Brand Distribution

```python
pd.crosstab(
    df_market["cluster"],
    df_market["Company"],
    normalize="index"
)
```

##### Touchscreen Ratio

```python
df_market.groupby("cluster")["is_touchscreen"].mean()
```

#### 4.Build Final Profile Table

```python
market_profile = (
    df_market.groupby("cluster")
    .agg({
        "Price": "mean",
        "ram_gb": "mean",
        "total_storage_gb": "mean",
        "cpu_tier_numeric": "mean",
        "is_touchscreen": "mean"
    })
    .round(2)
    .reset_index()
)

market_profile.columns = [
    "Cluster",
    "Avg Price",
    "Avg RAM (GB)",
    "Avg Storage (GB)",
    "Avg CPU Tier",
    "Touchscreen Ratio"
]

market_cluster_profile = market_profile.sort_values("Avg Price")

market_cluster_profile
```

```python
market_cluster_profile["Segment Name"] = [
    "Budget Segment",
    "Mid-Range Segment",
    "Premium Segment"
]
market_cluster_profile
```

#### conclusion

- The laptop market is strongly structured around pricing tiers. While technical specifications increase with price, premium features such as touchscreen capability significantly influence price positioning.
- This suggests that perceived user experience and design differentiation contribute meaningfully to market segmentation, beyond pure hardware performance metrics.
- The market-based clustering reveals a clear price-driven segmentation of laptops into three distinct economic tiers.

###### 0. Premium Segment

- Highest average price (~₹76K)

- Strong CPU tier and balanced RAM

- 100% touchscreen penetration

- This segment highlights premium user experience features, particularly touchscreen capability, which appears to command higher pricing even when raw specifications are comparable to the mid-range segment.

###### 1.Mid-Range Segment

- Moderate average price (~₹67K)

- Highest average storage capacity

- Strong CPU tier

- Minimal touchscreen presence

- This cluster represents performance-oriented productivity laptops, emphasizing storage and computing power over premium design elements.

###### 1.Budget Segment




- Lowest average price (~₹49K)

- Lowest RAM and CPU tier

- No touchscreen presence

- Moderate storage capacity

- This segment targets cost-sensitive users prioritizing affordability over advanced performance or premium features.

### compare hardware vs market segmentation:

```python
pd.crosstab(
    df_market["hardware_cluster"],
    df_market["cluster"],
    normalize="index"
)
```

### conclusion

- The cross-tabulation between hardware clusters and market clusters reveals that hardware identities do not map perfectly into pricing tiers.

- Certain hardware clusters are distributed across multiple price segments, indicating that similar technical configurations can exist at different price levels. This suggests that pricing is influenced not only by raw hardware specifications but also by factors such as brand positioning, design features, and perceived value.

- While market clustering shows strong price-based separation, hardware clustering reveals deeper technical differentiation that pricing alone does not fully capture.
