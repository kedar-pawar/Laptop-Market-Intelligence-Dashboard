```python
import pandas as pd
```

```python
df=pd.read_csv("../../02_Data_Management/04_Processed_Data/analytics_dataset.csv")
```

```python
df["Value_Score"] = df["Performance_Score"] / df["Price"]
```

## Rank Configurations

```python
df["Performance_Rank"] = df["Performance_Score"].rank(ascending=False)
df["Value_Rank"] = df["Value_Score"].rank(ascending=False)
```

```python
top_value = df.sort_values("Value_Score", ascending=False)
top_value.head(10)
```
