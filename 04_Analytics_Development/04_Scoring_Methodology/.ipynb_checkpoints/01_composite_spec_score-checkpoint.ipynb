```python
import pandas as pd
import numpy as np
from utils_preprocessing import preprocess
```

## data loading

```python
df=pd.read_csv("../../02_Data_Management/04_Processed_Data/analytics_dataset.csv")
df = preprocess(df)
df.info()
```

## Select Performance Features

```python
features = ["ram_gb", "ssd_gb", "cpu_tier", "clock_speed_ghz"]
perf_df = df[features]
```

## Normalize Features

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaled_values = scaler.fit_transform(perf_df)

scaled_df = pd.DataFrame(scaled_values, columns=features)
```

## Creating Weighted Scoring Formula

| Feature | Weight |
| ------- | ------ |
| cpu_tier     | 0.4    |
| clock_speed_ghz | 0.3    |
| RAM     | 0.2    |
| SSD     | 0.1    |


```python
weights = {
    "cpu_tier": 0.4,
    "clock_speed_ghz": 0.3,
    "ram_gb": 0.2,
    "ssd_gb": 0.1
}

scaled_df["Performance_Score"] = (
    scaled_df["cpu_tier"] * weights["cpu_tier"] +
    scaled_df["clock_speed_ghz"] * weights["clock_speed_ghz"] +
    scaled_df["ram_gb"] * weights["ram_gb"] +
    scaled_df["ssd_gb"] * weights["ssd_gb"]
)
```

```python
# Define performance features again
features = ["ram_gb", "ssd_gb", "cpu_tier", "clock_speed_ghz"]

# Define weights again
weights = {
    "cpu_tier": 0.4,
    "clock_speed_ghz": 0.3,
    "ram_gb": 0.2,
    "ssd_gb": 0.1
}
```

```python
df["Performance_Score"] = scaled_df["Performance_Score"]
```
